Captura de una noticia. Vamos a descargarnos una noticia de una pagina web y guardar su contenido en disco. 

```python
url = "https://elpais.com/ciencia/2024-11-29/la-ingeniosa-tecnica-de-moctezuma-y-su-banda-de-orcas-para-cazar-tiburones-ballena-en-mexico.html"

try:
    page = requests.get(url)
except:
    print("Error al abrir la URL")
    
```

```python
# Parseamos el html usando BeautifulSoup y lo guardamos en la variable "soup"
soup = BeautifulSoup(page.text, "html.parser")
```

```python
# Buscamos el <div> correspondiente y sacamos el contenido 
# (está en "inspeccionar en la pagina web, buscamos los <div>")
content = soup.find("div", {"data-dtm-region": "articulo_cuerpo"}) # el contenido entre {} está en el html de la página

if content:
    article = [] # lista para almacenar el texto de etiquetas <p> extraido, p significa párrafo en html
    for i in content.find_all("p"): # función para encontrar todos los párrafos y los retorna cómo una lista de elementos
        article.append(i.text) # para cada párrafo encontrado los añade a la big lista, pero solo el texto

    print("\n".join(article)) # junta todos los elementos dentro de la lista para que sea solo un texto, cada línea en la terminal corresponde a un párrafo
else:
    print("No se encontró el contenido del artículo.")
```

Vamos a extraer todos los links < a > dentro del artriculo. Busca los párrafos dentro de la etiqueta < div > con el cuerpo de la noticia que contengan un elemento de tipo enlace (< a >)

```python
content = soup.find("div", {"data-dtm-region": "articulo_cuerpo"})
links = content.find_all("a") # función para encontrar todos los links de content y los retorna cómo una lista de elementos

for link in links: # print los link
    print(f"Texto: {link.text}, Href: {link["href"]}") # .text para obtener el texto en el cual está el link y href es el hyperlink
```

Datos de la Wikipedia. Obtener las URL de las entradas en la wikipedia para todas las provincias de España y vamos a obtener de ellas sus datos básicos en forma de tabla.

```python
r = requests . get ("https://es.wikipedia.org/wiki/Provincia_(Espa%C3%B1a)") # extraemos el html de toda la pagina web
soup = BeautifulSoup(r.text, "html.parser") # cojemos el html y lo hacemos legible haciendo parsing
tabla = soup.find("table", class_="wikitable") # encontramos el html de la tabla entera

# todas las provincias de la tabla
provincias = tabla.find_all("tr") # extraemos las filas (tr que son table rows), le decimos que las coja todos y haga una lista

# accedemos a la segunda fila, porque la primera es el encabezado, prettify hace el html mas legible
print(provincias[1].td.prettify())
```

```python
# en la primera celda (table data o td) le decimos que coja todos los links <a>
provincias[1].td.find_all("a")
```

```python
enlaces = []  # Inicializa una lista vacía para almacenar los enlaces encontrados.

for provincia in provincias:  # Itera sobre cada fila (tr) de la tabla almacenada en la variable 'provincias'.
    if provincia.td:  # Verifica si la fila tiene al menos una celda (<td>). Esto evita errores si la fila está vacía.
        refs = provincia.td.find_all("a")  # Busca todos los enlaces (<a>) dentro de la primera celda (<td>) de la fila.
        for ref in refs:  # Itera sobre cada enlace encontrado en la celda.
            if ref.has_attr("title"):  # Verifica si el enlace tiene el atributo 'title'.
                enlaces.append(ref)  # Si el enlace tiene un atributo 'title', lo agrega a la lista 'enlaces'.
```

```python
enlaces
```

```python
enlaces_df = pd.DataFrame({  # Crea un DataFrame de pandas para estructurar los datos extraídos.
    "provincia": [e.attrs["title"] for e in enlaces],  # Extrae el atributo 'title' de cada enlace y lo almacena en la columna 'provincia'.
    "enlace": [e.attrs["href"] for e in enlaces]  # Extrae el atributo 'href' (URL) de cada enlace y lo almacena en la columna 'enlace'.
})
```

```python
enlaces_df["full_link"] = "https://es.wikipedia.org" + enlaces_df["enlace"]
enlaces_df
```

Ahora vamos a extraer en un data frame la información geográfica de la tabla de la barra lateral derecha (atributo de classe infobox)
Si inspeccionamos la estructura HTML veremos una serie de tags tr de las que cuelgan pares de tags th y td asociados. Capturaremos sus textos en dos listas: dato y valor, respectivamente

```python
# La variable 'r' ya contiene el HTML de la página web, por lo que no es necesario volver a hacer el request.
soup = BeautifulSoup(r.text, "html.parser")  # Parseamos el contenido HTML de la página con BeautifulSoup y lo guardamos en la variable 'soup'.

tabla = soup.find("table", class_="infobox")  # Buscamos la tabla con la clase "infobox" dentro del HTML. Esta tabla contiene información estructurada.

dato, valor = [], []  # Inicializamos dos listas vacías: 'dato' para almacenar los encabezados (th) y 'valor' para almacenar los datos asociados (td).
    #th son table headers (encabezados) y td son table data (celdas con data)
for t in tabla.find_all("tr"):  # Iteramos sobre todas las filas (<tr>) de la tabla.
    if t.th:  # Verificamos si la fila tiene un encabezado (<th>).
        if t.td:  # Verificamos si la fila también tiene un dato asociado (<td>).
            dato.append(t.th.text)  # Extraemos el texto del encabezado (<th>) y lo añadimos a la lista 'dato'.
            valor.append(" ".join([text for text in t.td.stripped_strings]))  
            # Extraemos el texto del dato (<td>), eliminamos espacios innecesarios con 'stripped_strings',
            # y unimos los fragmentos de texto en una sola cadena. Luego, lo añadimos a la lista 'valor'.

# Creamos un DataFrame de pandas con las listas 'dato' y 'valor'.
datos = pd.DataFrame({
    "Dato": dato,  # La columna "Dato" contiene los encabezados extraídos de la tabla.
    "Valor": valor  # La columna "Valor" contiene los datos asociados a cada encabezado.
})

datos  # Mostramos el DataFrame resultante.
```
